## FP64 64-bit (Double Precision) Floating Point Calculations
Many applications require higher-accuracy mathematical calculations. In these applications, data is represented by values that are twice as large (using 64 binary bits instead of 32 bits). These larger values are called double-precision (64-bit). Less accurate values are called single-precision (32-bit). Although almost all NVIDIA GPU products support both single- and double-precision calculations, the performance for double-precision values is significantly lower on most consumer-level GeForce GPUs.
